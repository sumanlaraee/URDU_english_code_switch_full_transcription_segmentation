```python
import whisper

model = whisper.load_model("large")
result = model.transcribe("/content/output_1min.wav", language="ur", task="translate", verbose=True)

# Save text + segments
with open("whisper_segments.txt", "w", encoding="utf-8") as f:
    for seg in result['segments']:
        f.write(f"[{seg['start']:.2f}–{seg['end']:.2f}] {seg['text']}\n")
```

```python
from pyannote.core import Segment

for ws in result["segments"]:
    whisper_seg = Segment(ws["start"], ws["end"])
    for turn in diarization.itertracks(yield_label=True):
        spk_seg, _, speaker = turn
        if whisper_seg.intersects(spk_seg):
            print(f"{speaker} [{ws['start']:.2f}–{ws['end']:.2f}]: {ws['text']}")
            break
```
